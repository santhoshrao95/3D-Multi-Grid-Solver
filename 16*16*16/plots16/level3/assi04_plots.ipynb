```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import csv
#from mpl_toolkits import mplot3d
import numpy as np
#import matplotlib.pyplot as plt
import math
file = ['mgoutput_1_3' , 'mgoutput_20_3','mgoutput_40_3']
for filename in file:
    with open(filename+'.csv') as inf:
        phi = []
   # y1 = []
        for line in csv.reader(inf):
            tx =  line
            tx = [float(i) for i in tx]
            phi.append(tx)
#print(phi)
    x = np.linspace(0, 1, 6)
    z = np.linspace(0, 1, 6)

    X, Z = np.meshgrid(x, z)
#phi = f(X, Z)
    y = np.array(phi)
#print(phi)
    ax = plt.axes(projection='3d')
    ax.plot_surface(X,Z,y[:], rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')
#ax.set_title('surface')


    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$z$')
    ax.set_zlabel('$phi$')
    ax.set_title(filename)
    plt.savefig(filename+'_3d.png')
    ax.view_init(30,-70)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

#from mpl_toolkits import mplot3d
import numpy as np
#import matplotlib.pyplot as plt
import math

file = ['mgoutput_1_3' , 'mgoutput_20_3','mgoutput_40_3']
for filename in file:
    with open(filename+'.csv') as inf:
        phi = []
   # y1 = []
        for line in csv.reader(inf):
            tx =  line
            tx = [float(i) for i in tx]
            phi.append(tx)
#print(phi)


    x = np.linspace(0.1, 1, 6)
    z = np.linspace(0.1, 1, 6)

    X, Z = np.meshgrid(x, z)
    y = np.array(phi)


    fig = plt.figure(figsize=(6,5))
    left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
    ax = fig.add_axes([left, bottom, width, height])

    cp = ax.contour(X, Z, y)
    ax.clabel(cp, inline=True,fontsize=10)

    ax.set_title(filename)
    plt.savefig(filename+'_contour.png')
```
